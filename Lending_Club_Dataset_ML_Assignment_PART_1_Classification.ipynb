#### Remove the columns which are highly correlated with other columns (to handle multi-collinearity issue)

```python
# Function to calculate correlation between numerical columns
def correlation(dataset, threshold):
    col_corr = []  # Set of all the names of correlated columns
    corr_matrix = dataset.corr() # Compute the correlation
    for i in range(len(corr_matrix.columns)): # Iterate through all binary column pairs
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # Append the both column names if the calculated correlation is more than threshold            		
                colname_1 = corr_matrix.columns[i]
                colname_2 = corr_matrix.columns[j]
                col_corr.append((colname_1,colname_2))
    return col_corr
```

```python
'''

One of the highly correlated numerical columns has to be removed from the above given column pairs. 
Explaination are as follows:

-> The total current balance (tot_cur_bal) is more important in credit risk because 
    it provides a better indication of the borrower's ability to pay back their debt. 
    A high credit limit (tot_hi_cred_lim) suggests that the borrower may be able to borrow more, 
    but it does not tell you how much of their current debt they are able to repay. 
    The total current balance provides insight into the borrower's financial situation and can be used to assess their creditworthiness.

-> Total bankcard high credit/credit limit (total_bc_limit) is more important in credit risk, because 
    it is indicative of a person's ability to manage credit responsibly.
    Total open to buy on revolving bankcards (bc_open_to_buy) is also important, as it helps lenders gauge how much debt a person can handle,
    but it is not as important as total bankcard high credit/credit limit since it does not directly measure a person's creditworthiness.

-> The ratio of total current balance to high credit/credit limit (bc_util) for all bankcard accounts is more important in credit risk
    than Percentage of all bankcard accounts > 75% of limit (percent_bc_gt_75).
    as it allows lenders to better assess an individual's ability to pay back their credit. 
    This ratio is a better indicator of an individual's financial health because it provides a more complete picture of the amount of debt
    they are carrying relative to the amount of credit available to them.
    This ratio can provide lenders with a better understanding of how much of a risk an individual poses, 
    and how likely they are to default on their loan.

-> The total current balance of all installment accounts (total_bal_il) is more important in credit risk, 
    as this reflects the current outstanding debt of the borrower. 
    This information is important in assessing the ability of the borrower to make regular payments,
    and the likelihood that they will default on the loan. 
    The total credit balance excluding mortgage (total_bal_ex_mort) is important in that it provides a picture of the borrower's overall creditworthiness and
    their total debt load, but it does not provide an indication of their ability to make payments in the near future.

-> The number of accounts opened in the past 12 months (num_tl_op_past_12m) is more important in assessing credit risk,
    as it provides an indication of the amount of debt that a customer has taken on in the past year. 
    The number of revolving trades opened in the past 12 months (open_rv_12m) is important as well, 
    since it can provide an indication of how often a customer is making payments on their debt.
    However, in terms of assessing overall credit risk, the number of accounts opened in the past 12 months is generally more important.

-> The number of revolving trades opened in the past 24 months (open_rv_24m) is more important in credit risk because it gives lenders a better
    indication of the dynamic nature of a borrower's credit profile. 
    Revolving trades are typically associated with credit cards, which means that they represent a borrower's ability to maintain their credit balance.
    This is more important than the number of trades opened in the past 24 months because
    it offers lenders a better indication of the borrower's ability to manage credit responsibly and pay off their debts in a timely manner.

-> The number of months since the borrower's last delinquency (mths_since_last_delinq) is more important in credit risk because 
    it is an indication of the borrower's recent payment history. 
    It can provide insight into whether the borrower is likely to default in the future. 
    Months since most recent revolving delinquency (mths_since_recent_revol_delinq) is important,
    but it is not as important as the number of months since the
    borrower's last delinquency because it does not provide as much information about the borrower's current credit situation.

-> (EXCEPTION) Generally, both number of public record bankruptcies (pub_rec_bankruptcies) and number of derogatory public records (pub_rec)
    are important in credit risk. 
    Public record bankruptcies provide an indication of how an individual or business has managed their finances in the past, 
    while derogatory public records provide an indication of a person or business's current financial status. 
    Both of these indicators can help a lender make an informed decision about a credit risk.
    'pub_rec_bankruptcies', 'pub_rec' => Keeping both

'''

correlated_columns_to_drop = ['num_sats',   # As it is correlating with two columns, ie, open_acc and num_op_rev_tl
'num_rev_tl_bal_gt_0',   # As it is correlating with two columns, ie, num_actv_bc_tl and num_actv_rev_tl
'tot_hi_cred_lim',
'bc_open_to_buy',
'percent_bc_gt_75',
'total_bal_ex_mort',
'num_rev_accts', # As it is correlating with two columns
'num_actv_bc_tl',  # As it is correlating with three columns, ie, num_actv_rev_tl, num_bc_sats and num_rev_tl_bal_gt_0
'avg_cur_bal',
'open_rv_12m',
'acc_open_past_24mths',
'mths_since_recent_revol_delinq',
'num_op_rev_tl',
'total_rev_hi_lim'
]
```

```python
# create a helper function clean up the emp_length column, assign 0 to NANs, and convert to numeric
def emp_length_converter(df, column):
    df[column] = df[column].str.replace('\+ years', '')
    df[column] = df[column].str.replace('< 1 year', str(0))
    df[column] = df[column].str.replace(' years', '')
    df[column] = df[column].str.replace(' year', '')
    df[column] = pd.to_numeric(df[column]) # ----- Numeric conversion ------
    df[column].fillna(value = 0, inplace = True)

# apply to training dataset
emp_length_converter(X_train_original, 'emp_length')

# Confirm the applied transformation
X_train_original['emp_length'].unique()
```

##### Convert data type for date column

```python
# convert date columns to datetime format and create a new column as a difference between 1st Jan 2022 and the respective date
def date_column_converter(df, column):
    # store current month
    jan_2022_date = pd.to_datetime('01-01-2022',format = "%b-%y",infer_datetime_format=True)
    # convert to datetime format
    df[column] = pd.to_datetime(df[column], format = "%b-%y", infer_datetime_format=True)
    # calculate the difference in months and add to a new column
    df['mths_since_' + column] = round(pd.to_numeric((jan_2022_date - df[column]) / np.timedelta64(1, 'M'))) # ----- Numeric conversion ------
    # make any resulting -ve values to be equal to the max date
    df['mths_since_' + column] = df['mths_since_' + column].apply(lambda x: df['mths_since_' + column].max() if x < 0 else x)
    # drop the original date column
    df.drop(columns = [column], inplace = True)


# apply 'date_column_converter' function on 'earliest_cr_line' (The month the borrower's earliest reported credit line was opened)
date_column_converter(df = X_train_original, column = 'earliest_cr_line')
```

```python
# define an empty dictionary to store chi-squared test results
chi2_check = {}

# loop over each column in the training set to calculate chi-statistic with the target variable
for column in X_train_cat:
    chi, p, dof, ex = chi2_contingency(pd.crosstab(y_train_original, X_train_cat[column]))
    chi2_check.setdefault('Feature',[]).append(column)
    chi2_check.setdefault('p-value',[]).append(round(p, 10))

# convert the dictionary to a dataframe to show final result
chi2_result = pd.DataFrame(data = chi2_check)
chi2_result.sort_values(by = ['p-value'], ascending = True, inplace = True)
chi2_result
```

```python
# since f_classif does not accept missing values, we will do a very crude imputation by replacing the missing values using median value from respective columns
X_train_num.fillna(X_train_num.median(), inplace = True)
# Calculate F Statistic and corresponding p values
F_statistic, p_values = f_classif(X_train_num, y_train_original)
# convert to a DF
ANOVA_F_table = pd.DataFrame(data = {'Numerical_Feature': X_train_num.columns.values, 'F-Score': F_statistic, 'p values': p_values.round(decimals=10)})
ANOVA_F_table.sort_values(by = ['F-Score'], ascending = False, inplace = True)
ANOVA_F_table.head(100)
```

#### Function for Stratified K-Cross Validation

```python
'''
This function performs the Stratified 5-Cross validation and will be used repeatedly for three selected models  
'''

def get_k_fold_cross_validation_scores(X,y,estimator,scoring_metric,model_name=None,n_splits=5,dataset_type=None,optimization_type_used=None):
  if dataset_type is not None:
    cv_scores = cross_val_score(estimator=estimator,\
                X = X,\
                y = y,\
                scoring=scoring_metric, # Scoring Meric such as accuracy, precision, recall, roc_auc etc
                cv=StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_SEED), # Cross validation using stratified folds
                n_jobs=-1)  
    return (
        "Cross Validation {} of {} Optimized {} model on {}: {:.3f} +/- {:.3f}".format(scoring_metric,optimization_type_used,model_name,dataset_type,np.mean(cv_scores),np.std(cv_scores)),
        np.mean(cv_scores), # Average Score
        np.std(cv_scores) # Score standard deviation
    )
```
